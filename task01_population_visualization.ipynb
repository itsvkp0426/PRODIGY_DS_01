```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/world_population.csv")
df.head()
```

```python
df.info()
```

```python
population_2022 = df[["Country/Territory", "2022 Population"]]
population_2022 = population_2022.dropna()
```

```python
plt.figure(figsize=(8,5))

sns.kdeplot(
    population_2022["2022 Population"],
    fill=True,
    color="#8172B2"
)

plt.title("Population Density Distribution (2022)", fontsize=13, fontweight="bold")
plt.xlabel("Population")
plt.ylabel("Density")

plt.tight_layout()
plt.savefig("visualization/population_distribution_2022.png", dpi=300)
plt.show()
```

```python
top10 = population_2022.sort_values(
    by="2022 Population",
    ascending=False
).head(10)

top10
```

```python
plt.figure(figsize=(9,5))

plt.barh(
    top10["Country/Territory"],
    top10["2022 Population"],
    color="#64B5CD"
)

plt.title("Top 10 Most Populated Countries (2022)", fontsize=13, fontweight="bold")
plt.xlabel("Population")
plt.ylabel("Country")

plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig("visualization/top10_population_2022.png", dpi=300)
plt.show()
```
